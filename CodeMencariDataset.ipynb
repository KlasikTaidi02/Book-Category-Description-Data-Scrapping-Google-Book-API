```python

from googleapiclient.discovery import build
import pandas as pd



df = pd.DataFrame(columns=['title', 'authors', 'categories','description'])

```

```python
books_service = build('books', 'v1', developerKey='AIzaSyAnEFDtCvyGFfL1lGa71HRVhV-lUNls-38')


collection = books_service.volumes()

start = 1


for i in range(1, 50):

    if i > 1:
        
        start += 40
    

        
    request = collection.list(q='Mathematic', maxResults=40 , printType='BOOKS' , orderBy='relevance', langRestrict='eng', startIndex=start )

   

    response = request.execute()

    print(response)

    if 'items' not in response:
        continue

    for item in response['items']:

        

        if 'authors' in item['volumeInfo']:
            title = item['volumeInfo']['title']
        else:
            title = 'N/A'

        if 'authors' in item['volumeInfo']:

            authors = item['volumeInfo']['authors']
        else:
            authors = 'N/A'

        if 'categories' in item['volumeInfo']:
            categories = item['volumeInfo']['categories']
        else:
            categories = 'N/A'

        if 'description' in item['volumeInfo']:
            description = item['volumeInfo']['description']
        else:
            description = 'N/A'

        df = pd.concat([df, pd.DataFrame({'title': [title], 'authors': [authors], 'categories': [categories], 'description': [description]})], ignore_index=True)
```
